```python
data = [
    {
        "iso_3": "CXR",
        "capital": "Flying Fish Cove",
        "capital_lat": "-10.4222",
        "capital_lng": "105.7895",
    },
    {
        "iso_3": "CCK",
        "capital": "West Island",
        "capital_lat": "-12.1643",
        "capital_lng": "96.8225",
    },
    {
        "iso_3": "LAO",
        "capital": "Vientiane",
        "capital_lat": "17.9750",
        "capital_lng": "102.6338",
    },
]
```

```python
import time
import polars as pl
from model import SessionLocal, Trade, Country, Product
session = SessionLocal()

for i in data:
    country = session.query(Country).filter(Country.iso_3 == i['iso_3']).first()
    if country:
        country.capital = i['capital']
        country.capital_lat = i['capital_lat']
        country.capital_lng = i['capital_lng']
        session.add(country)
    session.commit()

missing = session.query(Country).filter(Country.capital == None).all()
missing = [(x.name, x.iso_3) for x in missing]
print(missing)
print(len(missing))
```
